```python
from scipy import stats
import matplotlib.pyplot as plt
car_age=[4,10,20,5,2,11,1]
speed=[100,80,75,95,110,79,120]
m,c,r,p,std_err=stats.linregress(car_age,speed)
def fun(car_age):
    return (m*car_age)+c

p=list(map(fun,car_age))
plt.scatter(car_age,speed,marker='+',color='k')
plt.plot(car_age,p)
plt.show()
```
