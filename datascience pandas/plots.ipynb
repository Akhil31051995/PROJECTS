```python
#scatter plot

h=np.array([150,155,157,160,165,168,170,175,180,183,190])
w=np.array([55,56,59,75,45,80,75,65,66,59,74])

age=np.array([20,18,20,25,33,25,19,39,40,29,25])


plt.scatter(h,w,color='red')
plt.xlabel('height')
plt.ylabel('weight')


plt.scatter(h,age,color='green')
plt.xlabel('height')
plt.ylabel('age')


plt.title('scatter plot')
plt.xlabel('HEIGHT')
plt.ylabel('WEIGHT/AGE')

plt.show()
           
           
```

```python
plt.barh(h,w,color='red',height=1.5) #horizontal bar graph-----thickness given by height
plt.show()
plt.bar(h,age,width=1.2) #vertical bar---- thickness given by width
plt.show()
```

```python
#histogram

x=np.random.normal(170,10,250)      #with 10 std deviation around 170 generate 250 values

plt.hist(x,bins=5,color='red')      #histogram----------- bin size divides the no of intervals
plt.title('HISTOGRAM')
plt.xlabel('x-axis')
plt.ylabel('y-axis')


plt.show()
```

```python
#pie chart

y=np.array([25,35,60,54])
label=['mango','apple','banana','pineapple']
plt.pie(y,labels=label)
plt.show()
```

```python
y=np.array([25,35,60,54])
label=['mango','apple','banana','pineapple']
plt.pie(y,labels=label,startangle=90)           #angle of starting pie chart
plt.show()
```

```python
y=np.array([25,35,60,54])
label=['mango','apple','banana','pineapple']
plt.pie(y,labels=label,explode=[0.18,0.1,0,0])      #to detach from the chart
plt.show()
```

```python
y=np.array([25,35,60,54])
label=['mango','apple','banana','pineapple']
plt.pie(y,labels=label,shadow=True,colors=['g','r','k','c'])
plt.legend(loc='upper right',title='fruits')     # legend---marking the items ---loc---location of legends
plt.show()
```

```python
#group chart

x=np.array([2,3,6,5,8])
y1=np.array([5,8,6,9,7])
y2=np.array([8,5,7,8,9])

plt.bar(x-0.2,y1,width=0.4)             #x-0.2  first bar at a spacing 0.2 to left of each x value
plt.bar(x+0.2,y2,width=0.4)             #x+0.2  second bar at a spacing 0.2 to right of each x value
plt.show()
```

```python
x=np.array([2,3,6,5,8])
y1=np.array([5,8,6,9,7])
y2=np.array([8,5,7,8,9])
y3=np.array([7,5,6,3,2])

plt.bar(x-0.2,y1,width=0.2)             #x-0.2  first bar at a spacing 0.2 to left of each x value
plt.bar(x,y2,width=0.2)
plt.bar(x+0.2,y3,width=0.2)             #x+0.2  second bar at a spacing 0.2 to right of each x value
plt.legend(['a','b','c'])               # name of the bars given a,b ,c
plt.title('bar graph grouping')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xticks(x,['A','B','C','D','E'])         #change name/label of x values


plt.show()
```

```python
x=['A','B','C','D','E']
y1=np.array([5,8,6,9,7])
y2=np.array([4,5,4,8,6])
plt.bar(x,y1)
plt.bar(x,y2,bottom=y1)       #to stack the second values above the first y1 values
plt.show()
```

```python
x=['A','B','C','D','E']
y1=np.array([58,88,68,98,78])
y2=np.array([45,55,46,84,67])

y3=np.array([50,80,60,90,70])
y4=np.array([40,50,40,80,60])

plt.bar(x,y1)
plt.bar(x,y2,bottom=y1)
plt.bar(x,y3,bottom=y1+y2)
plt.bar(x,y4,bottom=y1+y2+y3)

plt.show()
```

```python
data={'player-name':['sachin','yuvraj','dhoni','kohli'],'matches':[10,23,50,75],'runs':[100,200,33,59],'wickets':[8,5,6,9]}

df=pd.DataFrame(data)        
df
```
