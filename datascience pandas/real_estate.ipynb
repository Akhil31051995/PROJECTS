```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Real estate.csv',usecols=['No','X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','Y house price of unit area'])
```

```python
df=df[0:50]
df
```

```python
df.isna().any()
```

```python
model=LinearRegression()
```

```python
x=df[['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores']]
```

```python
y=df['Y house price of unit area']
```

```python
model.fit(x,y)
```

```python
model.predict([[20,306,9]])
```
